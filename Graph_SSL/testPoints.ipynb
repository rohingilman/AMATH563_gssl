```python
from points import spiralPoints

import matplotlib.pyplot as plt
import numpy as np

from math import pi
```

```python
pts = spiralPoints(35,0,pi/2, 3)
pts = np.append(pts, spiralPoints(35,3*pi,7*pi/2, 3), axis = 0)
pts = np.append(pts, spiralPoints(35,4*pi,9*pi/2, 3), axis = 0)
```

```python
xs = pts[:,0]
ys = pts[:,1]

test = spiralPoints(10000,0,6*pi,0)

plt.scatter(test[:,0],test[:,1])
plt.scatter(xs,ys)
```
